```python
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import torch
import json

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large").to(device)
```

```python
text_caption={}
json_file_path="data/dev.jsonl"
i=int(0)
with open(json_file_path, 'r') as f:
    for line in f:
        entry = json.loads(line)
        if (i==10):
            with open("caption_dev.json", 'w') as f:
                  json.dump(text_caption, f)
            i=0
        else:
            i+=1
    # Iterate through each image entry in the JSON data
        # print(entry)
        # image_id = int(entry['id'])
        image_path = "data/"+entry['img']
        raw_image = Image.open(image_path).convert('RGB')
        # print(type(raw_image))
        inputs = processor(raw_image, return_tensors="pt").to(device)
        out = model.generate(**inputs,max_length=20)
        sentence=processor.decode(out[0], skip_special_tokens=True)
        text_caption[int(entry['id'])]={
            'sentence' : sentence,
        }
        print(sentence)
        print(entry['id'])
```

```python
with open("final_caption_dev.json", 'w') as f:
    json.dump(text_caption, f)
```
